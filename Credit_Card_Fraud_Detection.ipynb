```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('creditcard.csv')
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
rob_scaler = RobustScaler()

df['New_time'] = rob_scaler.fit_transform(df['Time'].values.reshape(-1,1))
df['New_amount'] = rob_scaler.fit_transform(df['Amount'].values.reshape(-1,1))
```

```python
df.drop(['Time', 'Amount'], axis=1, inplace=True)
```

```python
df['Class'].value_counts(normalize=True) * 100
```

```python
df['Class'].value_counts()
```

```python
from imblearn.under_sampling import TomekLinks
from imblearn.pipeline import Pipeline

from sklearn.preprocessing import PowerTransformer
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold, GridSearchCV, learning_curve
from sklearn.metrics import confusion_matrix, classification_report, brier_score_loss, precision_recall_curve
from sklearn.calibration import CalibratedClassifierCV
from sklearn.decomposition import PCA
```

```python
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
```

```python
X_raw = df.drop('Class', axis=1)
y_raw = df['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_raw, y_raw, test_size=0.2, random_state=42)
```

```python
pipeline2 = Pipeline(
    steps=[
        ('Scaler', RobustScaler()),
        ('Under_sampling', TomekLinks()),
        ('Transformer', PowerTransformer(method='yeo-johnson')),
        ('Classifier', MLPClassifier(hidden_layer_sizes=[100,100],
                                     random_state=42,
                                     activation='tanh'))
    ]
)
```

```python
pipeline2.fit(X_train, y_train)
```

```python
predict_p2 = pipeline2.predict(X_test)
```

```python
matrix_p2 = confusion_matrix(y_test, predict_p2)

sns.heatmap(matrix_p2, annot=True, cmap='Blues', cbar=False, fmt='d')
```

```python
print(classification_report(y_test, predict_p2, target_names=['Non-Fraud', 'Fraud']))
```

```python
precision_p2, recall_p2, thresholds_p2 = precision_recall_curve(y_test, predict_p2)

plt.figure()
plt.plot(precision_p2, recall_p2)
plt.xlabel('Precision')
plt.ylabel('Recall')
plt.title('Precision Recall Curve')
plt.show()
```
